```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("../Kaggle_Housing_Prices/train.csv")
pd.options.display.max_columns=None

data.head()
```

```python
## Clean dataset.  drop columns with a lot of na values.  20% na values is 292

clean_data=data.drop(['LotFrontage','Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'], axis=1)
clean_data.head()
```

```python
## MSSubClass is Categorical, not numeric.

clean_data['MSSubClass'] = clean_data['MSSubClass'].map({20: '1Story1946&newer', 30: '1Story1945&older', 40: '1StoryFinishedAttic', 45: '1-1/2StoryUnfinished', 50: '1-1/2StoryFinished', 60: '2Story1946&newer', 70: '2Story1945&older', 75: '2-1/2Story', 80: 'Split', 85: 'SplitFoyer', 90: 'Duplex', 120: '1StoryPUD1946&newer', 150: '1-1/2StoryPUD', 160: '2StoryPUD', 180: 'PUDMultiLevel', 190: '2FamilyConversion'})
```

```python
## Drop 1st floor sq ft and 2nd floor sq ft.  each individual
## line does not take into account relation to total sq ft 
## based on whether house has 1, 2 , 0r 3 floors.

clean_data=clean_data.drop(['1stFlrSF', '2ndFlrSF'], axis=1)
clean_data.head()
```

```python
binary_data= pd.get_dummies(clean_data)
binary_data=binary_data.fillna(0)
binary_data.head()
```

```python
X=binary_data.drop('SalePrice', axis=1)
y=binary_data[['SalePrice']].values.reshape(-1,1)

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, train_size=.8, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler=StandardScaler().fit(X_train)
```

```python
X_train_scaled=X_scaler.transform(X_train)
X_test_scaled=X_scaler.transform(X_test)
```

```python
model=LinearRegression()

model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
##model.predict(x_pred)
```

```python
binned_data=binary_data
```

```python
binned_data['year_bins']=  pd.cut(x=binned_data['YearBuilt'], bins=[1700, 1900, 1925, 1950, 1975, 2000, 2025])
binned_data.drop('YearBuilt', axis=1, inplace=True)
binned_data=pd.get_dummies(binned_data)
binned_data.head()
```

```python
Xw=binned_data.drop('SalePrice', axis=1)
yw=binned_data[['SalePrice']].values.reshape(-1,1)

print(Xw.shape)
print(yw.shape)
```

```python
Xw_train, Xw_test, yw_train, yw_test= train_test_split(Xw,yw, train_size=.8, random_state=42)

Xw_scaler=StandardScaler().fit(Xw_train)

Xw_train_scaled=Xw_scaler.transform(Xw_train)
Xw_test_scaled=Xw_scaler.transform(Xw_test)
```

```python
modelw=LinearRegression()

modelw.fit(Xw_train, yw_train)

print(modelw.score(Xw_train, yw_train))
print(modelw.score(Xw_test, yw_test))
##model.predict(x_pred)
```
